```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv


indoor_data = pd.read_csv("LAB5_indoors.csv")
outdoor_data = pd.read_csv("LAB5_outdoors.csv")


```

## Histograms

```python
plt.hist(indoor_data["PM2.5"], alpha = 0.5, bins = [0.5,1.5,2.5,3.5], color = "purple", label = "Indoor Data")
plt.hist(outdoor_data["PM2.5"], bins = [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5], alpha = 0.5, color = "red", label = "Outdoor Data" )
plt.title("PM2.5 Indoors vs Outdoors")
plt.legend()
plt.show()
```

Task 3a:
* The range for the indoor measurements is significantly smaller than for the outdoor measurements, therefore the boxes in the histograms are also much smaller.

```python
## sd of pm2.5

sm_PM25_i = np.std(indoor_data["PM2.5"])/np.sqrt(297)
sm_PM25_o = np.std(outdoor_data["PM2.5"])/np.sqrt(297)

meani = np.mean(indoor_data["PM2.5"])
meano = np.mean(outdoor_data["PM2.5"])

separation25 = (meano-meani)/sm_PM25_i

print(sm_PM25_i)
print("Separation =", separation25, "std. of the mean")
print("They are significantly different")
```

```python
## sd of pm10
sm_PM10_i = np.std(indoor_data["PM10"])/np.sqrt(297)
sm_PM10_o = np.std(outdoor_data["PM10"])/np.sqrt(297)

meani = np.mean(indoor_data["PM10"])
meano = np.mean(outdoor_data["PM10"])

separation10 = (meano-meani)/sm_PM10_i

print(sm_PM10_i)
print("Separation =", separation10, "std. of the mean")
print("They are significantly different")
```

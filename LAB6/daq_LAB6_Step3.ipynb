Task 3a:
* The range for the road measurements is significantly smaller for most variables than for the outdoor measurements, therefore the boxes in the histograms are also much smaller.

```python
## sd of pm2.5

sm_PM25_i = np.std(road_data["PM2.5"])/np.sqrt(297)
sm_PM25_o = np.std(outdoor_data["PM2.5"])/np.sqrt(297)

meani = np.mean(road_data["PM2.5"])
meano = np.mean(outdoor_data["PM2.5"])
stdi = np.std(road_data["PM2.5"])
stdo = np.std(outdoor_data["PM2.5"])

print("Road Mean: ", meani)
print("Outdoor Mean:", meano)
print("Road std.: ", stdi)
print("Outdoor std.:", stdo)

separation25 = (meano-meani)/sm_PM25_i

print(sm_PM25_i)
print("Separation =", separation25, "std. of the mean")
print("They are significantly different")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sc
import pandas as pd
import csv
```

```python
outdoors_uncut = pd.read_csv("LAB6_outdoors.csv")
outdoors = outdoors_uncut[127:]

road_uncut = pd.read_csv("LAB6_road.csv")
road = road_uncut[1:]
```

Below we wrote code for the Pearson Correlation method, but it broke down when we applied it because our data sets were non-linear. The hidden variable in all of these graphs is time.

```python
pearsons = []
mean_h = np.mean(road["Humidity"])
mean_t = np.mean(road["Temperature"])
for i in range(1,296):
    numerator = (road["Humidity"][i]-mean_h) * (road["Temperature"][i]-mean_t)
    denominator = np.sqrt((road["Humidity"][i]-mean_h)**2) * np.sqrt((road["Temperature"][i]-mean_t)**2)
    r = numerator/denominator
    pearsons.append(r)
```

Below, we tried to use the Spearman correlation to see if that made the fit better. It showed that our data was definitely correlated (P~10) but did not follow a linear correlation.

```python
plt.scatter(road["Temperature"], road["Humidity"])
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.title("Temperature vs Humidity")
plt.show()

sc.spearmanr(road["Temperature"], road["Humidity"])
```

```python
plt.scatter(road["Temperature"],road["Pressure"])
plt.xlabel("Temperature (C)")
plt.ylabel("Pressure")
plt.title("Temperature vs Pressure")
plt.show()
```

```python
plt.scatter(road["Temperature"],road["PM2.5"])
plt.xlabel("Temperature (C)")
plt.ylabel("Air Quality (PM2.5)")
plt.title("Temperature vs Air Quality (PM2.5)")
plt.show()
```

All of our graphs (except AQ), showed a non-linear correlation which would not be well described by a linear model. The air quality graph does not show anything significant because our sensor was not very sensitive and also was not exposed to our tested area long enough.

```python
## sd of pm2.5

sm_PM25_i = np.std(outdoor_data["PM2.5"])/np.sqrt(297)
sm_PM25_o = np.std(outdoor_data_Erick["PM2.5"])/np.sqrt(297)

meani = np.mean(outdoor_data["PM2.5"])
meano = np.mean(outdoor_data_Erick["PM2.5"])
stdi = np.std(outdoor_data["PM2.5"])
stdo = np.std(outdoor_data_Erick["PM2.5"])

print("Our Mean: ", meani)
print("Erick mean:", meano)
print("Our std.: ", stdi)
print("Erick std.:", stdo)

separation25 = (meano-meani)/sm_PM25_i

print( )
print("Separation =", separation25, "std. of the mean")
print("They are significantly different")
```

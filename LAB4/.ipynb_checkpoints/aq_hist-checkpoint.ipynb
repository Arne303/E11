```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('aq_data.csv')
# display(data)

plt.hist(data['PM2.5'], bins=3)
plt.xlabel(u"Concentration \u03bcg/L")
plt.ylabel('Frequency')
plt.title('Concentration of PM2.5')
plt.show()


plt.hist(data['PM10'], bins=3, color = 'green')
plt.xlabel(u"Concentration \u03bcg/L")
plt.ylabel('Frequency')
plt.title('Concentration of PM10')
plt.show()
```

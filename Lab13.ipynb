```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv

uncut_indoor = pd.read_csv("Loc1_session1_group1_2024-04-12_16-36-17.csv")

uncut_outdoor = pd.read_csv("Loc2_session1_group1_2024-04-12_16-45-33.csv")

indoor_data = uncut_indoor[2:]
outdoor_data = uncut_outdoor[2:]
```

```python
plt.plot(indoor_data["counts"], color = "purple", label = "Indoor")
plt.plot(outdoor_data["counts"], label = "Outdoor")
plt.xlabel("Time")
plt.ylabel("Counts")
plt.legend()
plt.title("Counts Outdoors and Indoors")
plt.show()
```

```python
plt.plot(indoor_data["Temperature"], color = "purple", label = "Indoor")
plt.plot(outdoor_data["Temperature"], label = "Outdoor")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.legend()
plt.title("Temperature Outdoors and Indoors")
plt.show()
```

```python
# b)

plt.hist(indoor_data["counts"], alpha = 0.5, color = "purple", label = "Indoor")
plt.hist(outdoor_data["counts"],alpha = 0.5, label = "Outdoor")
plt.xlabel("Counts")
plt.ylabel("Frequency")
plt.legend()
plt.title("Counts Outdoors and Indoors")
plt.show()
```
